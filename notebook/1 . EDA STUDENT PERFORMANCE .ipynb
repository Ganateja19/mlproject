```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Standardize column names (spaces -> underscores, lowercase, trim)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 2) Standardize gender values (so loc['female'] works)
df['gender'] = df['gender'].astype(str).str.strip().str.lower()

# 3) Create average if you haven't already
score_cols = [c for c in ['math_score', 'reading_score', 'writing_score'] if c in df.columns]
if 'average' not in df.columns:
    df['average'] = df[score_cols].mean(axis=1)

# 4) Rebuild gender_group with the correct numeric columns
gender_group = df.groupby('gender')[['average', 'math_score']].mean()

# 5) Plot
plt.figure(figsize=(10, 8))

X = ['Total Average', 'Math Average']

female_scores = [
    gender_group.loc['female', 'average'],
    gender_group.loc['female', 'math_score']
]

male_scores = [
    gender_group.loc['male', 'average'],
    gender_group.loc['male', 'math_score']
]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female')

plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average Marks by Gender", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights
- On average, **female students have a higher overall (total) average score** than male students.
- **Male students score higher in Mathematics** compared to female students.

- For every parental education level, **standard lunch** is more common than **free/reduced lunch**, indicating that the majority of students fall under the standard lunch category regardless of parental education.
- Overall, parental education appears to influence the **distribution of students** across categories, but the **dominant pattern remains consistent**: more students have *no test-prep* and receive *standard lunch* in each group.
